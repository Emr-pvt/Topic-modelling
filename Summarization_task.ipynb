```python
import torch
import json 
from transformers import T5Tokenizer, T5ForConditionalGeneration, T5Config
```

```python

model = T5ForConditionalGeneration.from_pretrained('t5-small')
tokenizer = T5Tokenizer.from_pretrained('t5-small')
device = torch.device('cpu')
```

You need to change the value below to select the abstract you want to summarize. We can automatize this by creating a for loop and a list that will contain all summaries.

```python
text =df['abstracts'][0]
text
```

```python
preprocess_text = text.strip().replace("\n","")

t5_prepared_Text = "summarize: " + preprocess_text
print ("original text preprocessed: \n", preprocess_text)

tokenized_text = tokenizer.encode(t5_prepared_Text, return_tensors="pt").to(device)
```

```python
# summmarize 
summary_ids = model.generate(tokenized_text,
                                    num_beams=4,
                                    no_repeat_ngram_size=2,
                                    min_length=30,
                                    max_length=100,
                                    early_stopping=True)

output = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

print ("\n\nSummarized text: \n",output)
```

```python
print("\n\nLongueur du text d'origine \n", len(text),"\n\nLongueur du text produit \n", len(output))
```
